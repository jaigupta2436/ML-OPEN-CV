```
 ## Linear Regression
def y_hat(theta,x):
    y_ = theta[0] + theta[1]*x
    return y_

def error(X,Y,theta):
    e = 0
    m = X.shape[0]
    for i in range(m):
        y_ = y_hat(theta,X[i])
        y = Y[i]
        e += (y_ - y)**2

    return e/m

def gradient(X,Y,theta):
    grad = np.zeros((2,))
    m = X.shape[0]
    for i in range(m):
        y_ = y_hat(theta,X[i])
        y = Y[i]
        grad[0] += (y_ - y)
        grad[1] += (y_ - y)*X[i]

    return 2*grad/m

def gradient_descent(X,Y):
    theta = np.zeros((2,))
    lr = 0.1
    error_list = []

    for i in range(50):
        e = error(X,Y,theta)
        error_list.append(e)
        grad = gradient(X,Y,theta)
        theta[0] = theta[0] - lr*grad[0]
        theta[1] = theta[1] - lr*grad[1]

    return theta,error_list  
```
